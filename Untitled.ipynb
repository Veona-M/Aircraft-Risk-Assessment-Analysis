# Aircraft Risk Assessment Analysis

## Introduction

In pursuit of portfolio diversification, our company is exploring expansion into the aviation sector, aiming to acquire and operate aircraft for commercial and private enterprises. However, successful entry into this industry requires a thorough understanding of associated risks. To support this strategic endeavor objectively, this analysis focuses on identifying aircraft with the lowest risk profiles, enabling a smooth and secure transition into the aviation market.

## Business understanding

The primary objective is to provide actionable insights that will assist the head of our new aviation division in selecting the safest aircraft. By determining which aircraft models present the lowest risk, we aim to mitigate potential hazards and establish a strong foundation for our aviation operations. 
This risk assessment will play a crucial role in guiding the purchasing decisions, ultimately contributing to the overall success and sustainability of our new aviation business.

```python
# Fucntion to categorize the purpose of flight into two categories namely: 'Personal' and 'Commercial'
# Assuming that any category not under personal is for commercial purposes.

def categorize(purpose):
    if purpose == 'Personal':
        return 'Personal'
    
    else:
        return 'Commercial'
```

#### Multivariate Analysis

```python
#Combine number of injuries regardless of severity
aviation_copy['Total_Injuries'] = aviation_copy['Total_Fatal_Injuries'] + aviation_copy['Total_Minor_Injuries'] + aviation_copy['Total_Serious_Injuries']
```

```python
#Create a function that accepts aviation data and a flight purpose as inputs, filters the data based on the specified purpose,
#and generates a plot illustrating the number of injuries related to the flight purpose.
def plot_injuries_summary(data, purpose):
    filtered_data = data[data['grouped_purpose_of_flight'] == purpose]
    summary = filtered_data[['Total_Injuries', 'Total_Uninjured']].sum().reset_index()
    summary.columns = ['Status', 'Count']

    plt.figure(figsize=(8, 4))
    sns.barplot(data=summary, x='Status', y='Count', palette='Set3')
    plt.title(f'Total Injured and Total Uninjured in {purpose} Flights')
    plt.xlabel('Status')
    plt.ylabel('Count')

    for idx, row in summary.iterrows():
        total = row['Count']
        percentage = (total / summary['Count'].sum()) * 100
        plt.text(idx, total, f'{percentage:.2f}%', ha='center', va='bottom')

    plt.show()
```

##### A. Personal Flight Accident Analysis

```python
#Call the function for personal flights plot
plot_injuries_summary(aviation_copy, 'Personal')
```

In personal flights, the number of uninjured individuals exceeds the number of injured ones.

```python
#Investigate the number of injured individuals for each make.

# Filter data for personal flights
personal_flights = aviation_copy[aviation_copy['grouped_purpose_of_flight'] == 'Personal']

# Select the top 20 aircraft makes used in personal flight accidents
top_20_makes = personal_flights['Make'].value_counts().head(20).index

# Filter the data for the top 20 makes
top_20 = personal_flights[personal_flights['Make'].isin(top_20_makes)]

# Group by make and sum the total injuries of individuals
injuries_by_make_personal = top_20.groupby('Make')[['Total_Injuries']].sum().sort_values(by='Total_Injuries', ascending=False)

# Display the counts for injured and uninjured individuals for each aircraft make
plt.figure(figsize=(14,6))
sns.barplot(data=injuries_by_make_personal, x= injuries_by_make_personal.index, y= 'Total_Injuries', palette='Set3')
plt.title("Injuries for top 20 Aircraft makes from Personal Flights")
plt.xlabel("Aircraft Make")
plt.ylabel("Number of Injuries")
plt.xticks(rotation= 60)
plt.show()
```

The highest number of injuries was observed in Cessna, with 11,973 injuries, closely followed by Piper, with 8,778 injuries.

```python
#Investigate the number of uninjured individuals for each make.

# Group by make and sum the total uninjured individuals
uninjured_by_make_personal = top_20.groupby('Make')[[ 'Total_Uninjured']].sum().sort_values(by='Total_Uninjured', ascending=False)

# Display the counts for uninjured individuals for each aircraft make
plt.figure(figsize=(14,6))
sns.barplot(data=uninjured_by_make_personal, x= uninjured_by_make_personal.index, y= 'Total_Uninjured', palette='Set3')
plt.title("Number of Uninjured for top 20 Aircraft makes from Personal Flights")
plt.xlabel("Aircraft Make")
plt.ylabel("Number of Uninjured")
plt.xticks(rotation= 60)
plt.show()
```

The plot shows which aircraft make has the highest number of uninjured individuals in personal flight accidents. Cessna leads with the highest number of uninjured individuals at 18,974, followed by Piper at 10,965 and Beech at 3,459.

```python
# Check the ratio of injured to uninjured individuals for different aircraft makes  

# Group personal flights by aircraft make
personal_flights_make = top_20.groupby('Make')

# Calculate sum of total injured and total uninjured for each aircraft make
sum_injured_uninjured = personal_flights_make[['Total_Injuries', 'Total_Uninjured']].sum()

# Calculate ratio of injured to uninjured
sum_injured_uninjured['Injured_to_Uninjured_Ratio_personal'] = (sum_injured_uninjured['Total_Injuries'] / sum_injured_uninjured['Total_Uninjured'])

# Display the result in a pivot table

pivot_table_ratio = sum_injured_uninjured.pivot_table(index='Make', values=['Total_Injuries', 'Total_Uninjured', 'Injured_to_Uninjured_Ratio_personal'])
pivot_table_ratio_sorted = pivot_table_ratio.sort_values(by='Injured_to_Uninjured_Ratio_personal', ascending=False)
pivot_table_ratio_sorted
```

The ratios provide insight into the safety performance of different aircraft makes. A lower ratio indicates a better outcome in terms of passenger safety during accidents.

Piper and Cessna, despite having a high number of total injuries 8,778 and 11,973, respectively, have lower ratios of injured to uninjured individuals (0.8005478 and 0.631921), suggesting a relatively higher number of uninjured individuals in their accidents.

```python
# Merge the pivot_table_ratio_sorted with aviation_copy
aviation_copy = pd.merge(aviation_copy, pivot_table_ratio_sorted['Injured_to_Uninjured_Ratio_personal'], on='Make', how='left')
```

```python
# Investigate the total number of injuries based on the combination of aircraft make and the extent of aircraft damage in personal flights. 

# Group by aircraft make and damage, and sum the total injuries
damage_by_make_personal = top_20.groupby(['Make', 'Aircraft_damage'])['Total_Injuries'].sum().reset_index()

# Create a pivot table
pivot_table_damage_personal = damage_by_make_personal.pivot_table(index='Make', columns='Aircraft_damage', values='Total_Injuries', fill_value=0)

# Display the pivot table
pivot_table_damage_personal

#Plot
pivot_table_damage_personal.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Aircraft Damage by Make in Personal Flights')
plt.xlabel('Aircraft Make')
plt.ylabel('Total Injuries')
plt.xticks(rotation=75)
plt.legend(title='Aircraft Damage')
plt.show()
```

The plot provides a quick comparison of injury counts among different damage categories within each aircraft make. The more severe the damage, the higher the number of injuries.

```python
#comparing the number of personal accidents caused by each engine type

# Extract data for engine types
engine_type_data = aviation_copy[['Make', 'Engine_Type', 'Total_Injuries']]

# Group data by aircraft make and engine type and calculate the sum for total injuries
injuries_by_make_and_engine_type = engine_type_data.groupby(['Make', 'Engine_Type'])['Total_Injuries'].sum().reset_index()

# Select the top 20 aircraft makes
top_20_makes_engine = injuries_by_make_and_engine_type.sort_values(by='Total_Injuries', ascending=False).head(20)['Make']

# Filter the data for the top 20 makes
injuries_by_make_and_engine_type_top_20 = injuries_by_make_and_engine_type[injuries_by_make_and_engine_type['Make'].isin(top_20_makes_engine)]

# Visualize comparison
plt.figure(figsize=(14, 8))
sns.barplot(data=injuries_by_make_and_engine_type_top_20, x='Make', y='Total_Injuries', hue='Engine_Type', palette='Set3')
plt.title("Comparison of Total Injuries in Personal Flight Accidents by Top 20 Aircraft Makes and Engine Types")
plt.xlabel("Aircraft Make")
plt.ylabel("Total Injuries")
plt.xticks(rotation=90)
plt.legend(title='Engine Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Aircrafts with reciprocating engine have a high number of total injuries as is evident in Cessna, Piper, Beech. However boeing aircrafts with turbo fan engine have higher number of injuries in their class.

## Conclusion

Based on the analysis conducted, it is evident that several key insights emerge regarding passenger safety and aircraft damage severity. Firstly, the ratio of injured to uninjured passengers serves as a crucial indicator, with higher ratios signaling lower levels of safety for passengers. Secondly, there exists a direct correlation between the severity of aircraft damage and the number of injured passengers, emphasizing the importance of effective maintenance and safety measures. Additionally, the choice of engine type significantly influences injury counts, with distinct patterns observed across different aircraft models. Reciprocating engines are associated with higher injury counts, while Turbofan engines exhibit elevated injury counts across both personal and commercial aircraft categories. These findings underscore the importance of considering factors such as aircraft damage severity and engine type in assessing passenger safety and injury risk, informing decision-making processes aimed at the purchase of new aircraft and prioritizing safety measures to safeguard passenger well-being.

## Recommendation

Based on the analysis of aircraft safety data, the following recommendations are made to guide investment decisions in the aviation sector:

1. Investment in Aircraft with Low Injury Ratios:

Personal Use: Prioritize investment in aircraft models with a low injured-to-uninjured passenger ratio, as they demonstrate superior passenger safety outcomes. Specifically, models such as Piper and Cessna are recommended.
Commercial Use: For commercial operations, aircraft models such as Boeing and Cessna are advisable due to their favorable safety records.

2. Engine Type Considerations:

Personal Use: It is advisable to avoid aircraft with reciprocating engines, particularly for models such as Cessna and Piper, as they are correlated with higher injury counts. Instead, consider opting for aircraft equipped with turboprop engines with a better safety profile.
Commercial Use: For commercial aircraft, it is recommended to avoid turbofan engines in models such as Boeing and McDonnell Douglas. Opting for aircraft with reciprocating or turbojet engines is advisable, as these engines are associated with minimal injuries.
By focusing on aircraft with proven safety records and selecting appropriate engine types, the company can enhance passenger safety, mitigate operational risks, and make informed investment decisions in the aviation industry.

3. From the analysis, it is clear that aircraft are susceptible to accidents; therefore, the organization should implement robust safety measures and maintenance protocols to ensure smooth operations and high aircraft performance.
